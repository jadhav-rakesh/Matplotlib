```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, "o", color="purple")
```

```python
rng = np.random.default_rng(0)

for marker in ["o", ".", ",", "x", "+", "v", "^", "<", ">", "s", "d"]:
    plt.plot(rng.random(2), rng.random(2), marker, color="purple",
             label="marker='{0}".format(marker))
    plt.legend(numpoints=1, fontsize=13, loc="right")
    plt.xlim(0, 1.0)
```

```python
plt.plot(x, y, "-ok")
```

```python
plt.plot(x, y, "-p", color="red",
         markersize=15, linewidth=4,
         markerfacecolor="green",
         markeredgecolor="yellow",
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

```python
plt.scatter(x, y, marker="o")
```

```python
rng = np.random.default_rng(42)
x = rng.normal(size=100)
y = rng.normal(size=100)

colors = rng.random(100)
sizes = 1000 * rng.random(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.colorbar()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
features= iris.data.T

plt.scatter(features[0], features[1], alpha=0.4,
            s=100*features[3], c=iris.target, cmap="viridis")

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy + np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt=".k")
```

```python
plt.errorbar(x, y, yerr=dy, fmt="o", color="black",
             ecolor="lightgray", elinewidth=3, capsize=0)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor

model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, dyfit = gp.predict(xfit[:, np.newaxis], return_std=True)
```

```python
plt.plot(xdata, ydata, "or")
plt.plot(xfit, yfit, "-", color="gray")
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color="gray", alpha=0.2)
plt.xlim(0, 10)
```
