```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

```python
plt.style.available
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
plt.plot(x, np.sin(x - 0), color="blue")        #name
plt.plot(x, np.sin(x-1), color="g")             #color code(rbcmyk)
plt.plot(x, np.sin(x-2), color="0.75")          #grayscale(0-1)
plt.plot(x, np.sin(x-3), color="#FFDD44")       #hex code(RRGGBB 00 - FF)
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3)) #RGB (0- 1)
plt.plot(x, np.sin(x-5), color="chartreuse")    #HTML color names
```

```python
plt.plot(x, x + 0, linestyle="solid")
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':')
```

```python
plt.plot(x, x + 0, "-g")
plt.plot(x, x + 1, "--c")
plt.plot(x, x + 2, "-.k")
plt.plot(x, x + 3, ":r")
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# axis to be displayed in reverse

plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis("tight")
```

```python
plt.plot(x, np.sin(x))
plt.axis("equal")
```

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), "-y", label="sin(x)")
plt.plot(x, np.cos(x), ":r", label="cos(x)")
plt.axis("equal")

plt.legend()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel="x", ylabel="sin(x)",
       title="A Simple Plot")
```
